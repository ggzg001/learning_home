## 3. pandas中的None与NaN

### 1) pandas中None与np.nan都视作np.nan

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

创建DataFrame

```python
df = DataFrame(data=np.random.randint(10,50,size=(8,8)))
df
```

```python
#将某些数组元素赋值为nan
```

```python
df.iloc[1,3] = None
df.iloc[2,2] = None
df.iloc[4,2] = None
df.iloc[6,7] = np.nan
```

```python
df
```

### 2) pandas处理空值操作

- ``isnull()``
- ``notnull()``
- ``dropna()``: 过滤丢失数据
- ``fillna()``: 填充丢失数据

```python
df.isnull()
```

```python
#创建DataFrame，给其中某些元素赋值为nan

df.notnull().all(axis=1)     #notnull(all)   isnull(any)
```

```python
df.loc[df.notnull().all(axis=1)]
```
